```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result-01.csv')

starting_point = 40
ending_point = 60
plt.errorbar(df['beta'][starting_point:ending_point], df['energy'][starting_point:ending_point], yerr=df['standard_deviation'][starting_point:ending_point], color='black', marker='o', ecolor='gray', capsize=2)
plt.ylabel("Energy")
plt.xlabel(r"$\beta$")
plt.grid()
plt.show()
```
