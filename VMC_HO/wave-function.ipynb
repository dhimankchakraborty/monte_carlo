```python
from sage.all import *
```

```python
var('x beta')
assume(beta > 0)

phi = exp(-beta * x**2)

prob_density = phi**2

integral_val = integral(prob_density, x, -infinity, infinity)

normalization_factor = 1 / sqrt(integral_val)

print(f"Integral of |phi|^2: {integral_val}")
print(f"Normalization Factor A: {normalization_factor}")
```
