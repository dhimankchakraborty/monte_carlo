```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
total_points_count = 10**9

rng = np.random.default_rng()
rnd_coord_arr = rng.random((total_points_count, 2))
```

```python
points_count_inside_circle = 0
for i in tqdm(range(total_points_count)):
    if (rnd_coord_arr[i][0]**2 + rnd_coord_arr[i][1]**2) <= 1:
        points_count_inside_circle += 1

print(f"Estimated value of Pi: {4 * points_count_inside_circle / total_points_count}")
print(f"Error in the esimated value of Pi: {(((4 * points_count_inside_circle / total_points_count) - np.pi) / np.pi * 100):.5f}%")
```
