```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv("results.csv")
```

```python
plt.figure() 
plt.semilogx(df["Points"], df["Estimated_Pi"], marker='o') 
plt.xlabel("Number of Points (N)") 
plt.ylabel("Estimated Pi") 
plt.title("Monte Carlo Estimation of Pi") 
plt.grid(True) 
plt.show()
```

```python
theory = 1 / np.sqrt(df["Points"])
plt.figure()
plt.loglog(df["Points"], np.abs(df["Error(%)"]), label="Simulation", marker='o')
plt.loglog(df["Points"], theory * 100, '--', label="~1/sqrt(N)")
plt.xlabel("Number of Points (N)")
plt.ylabel("Error (%)")
plt.title("Monte Carlo Error Scaling")
plt.legend()
plt.grid(True, which="both")
plt.show()
```
